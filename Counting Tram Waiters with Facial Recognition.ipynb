# Counting Tram Waiters with Facial Recognition

```python
%matplotlib notebook
import tramface
import tramset
import matplotlib.pyplot as plt

camera_ip = "http://158.108.253.207:8888"
api_ip = "https://kutram.herokuapp.com"

stop = tramset.Stop(1, api_ip)
```

```python
stop.update_waiting(3)
```

```python
tramface.snap(camera_ip)
fig, ax = plt.subplots()
faces, encodings = tramface.get_faces("snap.jpg", ax=ax)
fig.show()
n_faces = len(faces)
stop.update_waiting(n_faces)
```
